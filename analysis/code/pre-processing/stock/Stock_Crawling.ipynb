### HYBE

```python
# Í∏∞Ïà†Ï†Å ÏßÄÌëú Í≥ÑÏÇ∞

# MACD Í≥ÑÏÇ∞
stock_hybe['MACD'], stock_hybe['Signal'], _ = talib.MACD(stock_hybe['Close'])

# PSAR Í≥ÑÏÇ∞
stock_hybe['PSAR'] = talib.SAR(stock_hybe['High'], stock_hybe['Low'])

# Bollinger Bands (BB) Í≥ÑÏÇ∞
# *** BBÎäî upper, middle, lower ÏÑ∏ Ï¢ÖÎ•òÎ°ú Íµ¨ÏÑ±Îê®
bb_period = 20
stock_hybe['upper'], stock_hybe['middle'], stock_hybe['lower'] = talib.BBANDS(stock_hybe['Close'], timeperiod=bb_period)

# Stochastic Oscillator (SO) Í≥ÑÏÇ∞
# *** SOÎäî SlowK, SlowD Îëê Ï¢ÖÎ•òÎ°ú Íµ¨ÏÑ±Îê®
slowk_period = 14
slowd_period = 3
stock_hybe['SlowK'], stock_hybe['SlowD'] = talib.STOCH(stock_hybe['High'], stock_hybe['Low'], stock_hybe['Close'], fastk_period=slowk_period, slowd_period=slowd_period)

# Rate of Change (ROC) Í≥ÑÏÇ∞
roc_period = 12
stock_hybe['ROC'] = talib.ROC(stock_hybe['Close'], timeperiod=roc_period)

# On-Balance Volume (OBV) Í≥ÑÏÇ∞
stock_hybe['OBV'] = talib.OBV(stock_hybe['Close'], stock_hybe['Volume'])

# Force Index (FI) Í≥ÑÏÇ∞
stock_hybe['FI'] = (stock_hybe['Close'] - stock_hybe['Close'].shift(1)) * stock_hybe['Volume']

# Í≤∞Í≥º ÌôïÏù∏
print(stock_hybe)
```
