# (2) YouTube 데이터 변수 정리
- (1) 새로운 변수인 log_like, view_log_like를 생성함.
    - log_like: like변수에 로그변환 취한 변수
    - view_log_like: view 변수와 log_like 변수를 곱한 변수
    - *YouTube 데이터프레임에는 '0'이 없으므로 로그변환함수는 np.log()를 씀.*
- (2) date 변수의 형태 통일
    - youtube_sm만 날짜 형식이고 나머지 3개의 데이터프레임은 날짜+시간 형식임.
    - 이들을 날짜 형식으로 통일함.
- (3) 필요없는 변수들은 제거함.

```python
#log_like 변수 생성
youtube_hybe["log_like"] = np.log(youtube_hybe["like"])
youtube_jyp["log_like"] = np.log(youtube_jyp["like"])
youtube_sm["log_like"] = np.log(youtube_sm["like"])
youtube_yg["log_like"] = np.log(youtube_yg["like"])
```

```python
#view_log_like
youtube_hybe["view_log_like"] = youtube_hybe["view"] * youtube_hybe["log_like"]
youtube_jyp["view_log_like"] = youtube_jyp["view"] * youtube_jyp["log_like"]
youtube_sm["view_log_like"] = youtube_sm["view"] * youtube_sm["log_like"]
youtube_yg["view_log_like"] = youtube_yg["view"] * youtube_yg["log_like"]
```

```python
#하이브(hybe)의 데이터로 변수가 잘 생성되었는지 확인
youtube_hybe.head()
```

```python
#date 변수에서 datetime 형태를 date 형태로 바꿔주는 함수 생성
#split() 함수를 이용해 문자열을 분리함
def datetime2date(df):
    for i in range(len(df)):
        df["date"][i] = df["date"][i].split(" ")[0]
```

```python
#youtube_hybe, youtube_jyp, youtube_yg에 대해서 datetime2date() 함수를 적용
datetime2date(youtube_hybe)
datetime2date(youtube_jyp)
datetime2date(youtube_yg)
```

```python
#하이브(hybe)의 데이터로 date 변수가 잘 변경되었는지 확인
youtube_hybe.head()
```

```python
#필요없는 변수 제거
youtube_hybe = youtube_hybe.drop(["categoryId","dislike","comment","tags"], axis=1)
youtube_jyp = youtube_jyp.drop(["categoryId","dislike","comment","tags"], axis=1)
youtube_sm = youtube_sm.drop(["categoryId","dislike","comment","tags"], axis=1)
youtube_yg = youtube_yg.drop(["categoryId","dislike","comment","tags"], axis=1)
```

## (3-2) 소속사별 날짜 기준 데이터프레임 생성
각 소속사의 유튜브와 네이버 데이터를 날짜 기준으로 결합한 데이터프레임을 생성함.
<br>**<span style = "color:Red">(주의!!) 생성된 데이터프레임의 행 갯수는 181(2023.01 ~ 2023.06의 날짜수)이어야 함!</span>**

<br>**<데이터프레임에 들어갈 변수 목록>**
- (1) view_log_like_sum: 날짜별 view_log_like의 합계
- (2) view_log_like_avg: 날짜별 view_log_like의 평균
- (3) count: 날짜별 업로드된 동영상 개수
- (4) keyword: 날짜별 키워드 데이터

**하이브(HYBE)**

```python
#네이버 데이터를 기준으로 최종 데이터프레임 틀 생성
trend_hybe = pd.DataFrame({"date":naver_hybe["date"], "keyword":naver_hybe["ratio_sum"]})

#유튜브 데이터로 만든 부수적인 데이터프레임
view_log_like_sum = pd.DataFrame({"date":list(youtube_hybe[["date","view_log_like"]].groupby("date").sum().index), "view_log_like_sum":youtube_hybe[["date","view_log_like"]].groupby("date").sum()["view_log_like"]}).reset_index(drop=True)
view_log_like_avg = pd.DataFrame({"date":list(youtube_hybe[["date","view_log_like"]].groupby("date").mean().index), "view_log_like_avg":youtube_hybe[["date","view_log_like"]].groupby("date").mean()["view_log_like"]}).reset_index(drop=True)
count = pd.DataFrame({"date":list(youtube_hybe.groupby("date").count()["videoId"].index), "count":youtube_hybe.groupby("date").count()["videoId"]}).reset_index(drop=True)

#데이터프레임 병합
trend_hybe = pd.merge(trend_hybe, view_log_like_sum, on="date", how="left")
trend_hybe = pd.merge(trend_hybe, view_log_like_avg, on="date", how="left")
trend_hybe = pd.merge(trend_hybe, count, on="date", how="left")

#결측치는 '0'으로 변환
trend_hybe = trend_hybe.fillna(0)

trend_hybe
```

## (3-3) 트렌드 변수 생성


**하이브(HYBE)**

```python
trend_list = []

for i in range(len(trend_hybe)):
    if trend_hybe["count"][i] != 0: #당일 업로드된 동영상 개수가 '0'이 아닌 경우
        trend = np.log(np.log(trend_hybe["count"][i]+np.exp(1)) * trend_hybe["view_log_like_avg"][i] * trend_hybe["keyword"][i])
        trend_list.append(trend)
    else: #당일 업로드된 동영상 개수가 '0'인 경우
        trend = np.log(np.log(trend_hybe["count"][i]+np.exp(1)) * 1 * trend_hybe["keyword"][i])
        trend_list.append(trend)

trend_hybe["trend"] = trend_list
trend_hybe
```
