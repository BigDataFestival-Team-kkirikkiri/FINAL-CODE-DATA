ì´ ë‹¨ê³„ì—ì„œëŠ” ìš”ì²­ëŸ‰ì´ ë§ì•„ì„œ API ìš”ì²­ì´ ëŠê¸¸ ìˆ˜ ìˆìŒ!!

### (1ë²ˆ) JYP Entertainment
ì œì´ì™€ì´í”¼(JYP)

```python
item_jyp = pd.read_csv("../../../data/YouTube/item_jyp.csv")
item_jyp
```

```python
titles = [] #ë™ì˜ìƒ ì œëª©
ids = [] #ë™ì˜ìƒ id
dates = [] #ë™ì˜ìƒ ì—…ë¡œë“œ ë‚ ì§œ
category_ids = [] #ë™ì˜ìƒ ì¹´í…Œê³ ë¦¬ id
views = [] #ì¡°íšŒìˆ˜
likes = [] #ì¢‹ì•„ìš” ìˆ˜
dislikes = [] #ì‹«ì–´ìš” ìˆ˜
comments = [] #ëŒ“ê¸€ ìˆ˜
tags = [] #íƒœê·¸ ì •ë³´

hours = [] #ë™ì˜ìƒ ì¬ìƒê¸¸ì´(ì‹œ)
mins = [] #ë™ì˜ìƒ ì¬ìƒê¸¸ì´(ë¶„)
secs = [] #ë™ì˜ìƒ ì¬ìƒê¸¸ì´(ì´ˆ)

#ë¦¬ìŠ¤íŠ¸ í˜•íƒœì˜ tags ì •ë³´ë¥¼ ì¼ë ¨ì˜ ë¬¸ìì—´ë¡œ ë³€í™˜í•´ì£¼ëŠ” í•¨ìˆ˜
def Tags2Strings(tags_list):
    
    string = ""
    for i in range(len(tags_list)):
        string = string + tags_list[i] + "|"
    string = string[:-1] #ë¬¸ìì—´ ë§¨ ëì— ë¶™ì€ "|"ì€ ì—†ì• ë²„ë¦¼
    
    return string

#ë°ì´í„° ì¶”ì¶œ
for i in range(len(item_jyp)): #JYP Entertainment
    request = youtube.videos().list(
        id = item_jyp["video_id"][i],
        part = "snippet,contentDetails,statistics"
    )
    video_response = request.execute()
    
    if video_response["items"] == []: #ë™ì˜ìƒ ì •ë³´ê°€ ì—†ì„ ê²½ìš° '-'ê°’ ì…ë ¥
        titles.append("-")
        ids.append("-")
        dates.append("-")
        category_ids.append("-")
        views.append("-")
        likes.append("-")
        dislikes.append("-")
        comments.append("-")
        tags.append("-")
        hours.append("-")
        mins.append("-")
        secs.append("-")
    
    else:
        titles.append(video_response["items"][0]["snippet"]["title"])
        ids.append(video_response["items"][0]["id"])
        dates.append(video_response["items"][0]["snippet"]["publishedAt"].split("T")[0])
        category_ids.append(video_response["items"][0]["snippet"]["categoryId"])
        
        #views ë°ì´í„°
        if "viewCount" in video_response["items"][0]["statistics"].keys():
            views.append(video_response["items"][0]["statistics"]["viewCount"])
            
        else:
            views.append("-")
        
        #comments ë°ì´í„°
        if "commentCount" in video_response["items"][0]["statistics"].keys():
            comments.append(video_response["items"][0]["statistics"]["commentCount"])
            
        else:
            comments.append("-")
        
        #likes ë°ì´í„°
        if "likeCount" in video_response["items"][0]["statistics"].keys():
            likes.append(video_response["items"][0]["statistics"]["likeCount"])
            
        else: #likeCountê°€ ì—†ëŠ” ê²½ìš°ë„ ìˆìŒ
            likes.append("-")
        
        #dislikes ë°ì´í„°
        if "dislikeCount" in video_response["items"][0]["statistics"].keys():
            dislikes.append(video_response["items"][0]["statistics"]["dislikeCount"])
            
        else:
            dislikes.append("-")
        
        #tags ë°ì´í„°
        if "tags" in video_response["items"][0]["snippet"].keys():
            tags_string = video_response["items"][0]["snippet"]["tags"]
            tags.append(Tags2Strings(tags_string))
            
        else: #tagsê°€ ì—†ëŠ” ê²½ìš°ë„ ìˆìŒ(shorts í˜•ì‹ì˜ ì˜ìƒë“¤)
            tags.append("-")
        
        #hour, min, sec ë°ì´í„°
        duration = re.findall(r'\d+', video_response["items"][0]["contentDetails"]["duration"])
        if len(duration) == 3:
            hours.append(duration[0])
            mins.append(duration[1])
            secs.append(duration[2])
        elif len(duration) == 2:
            hours.append("-")
            mins.append(duration[0])
            secs.append(duration[1])
        else:
            hours.append("-")
            mins.append("-")
            secs.append(duration[0])
        
        
video_df = pd.DataFrame({"title": titles, "videoId": ids, "date": dates, "categoryId": category_ids,
                        "view": views, "like": likes, "dislike": dislikes, "comment": comments, "tags": tags,
                        "hour": hours, "min": mins, "sec": secs})
video_df
```

```python
#video_dfë¥¼ csvíŒŒì¼ë¡œ ì €ì¥
video_df.to_csv("../../../data/YouTube/video_jyp.csv", index=False)
```

```python
titles = [] #ë™ì˜ìƒ ì œëª©
ids = [] #ë™ì˜ìƒ id
dates = [] #ë™ì˜ìƒ ì—…ë¡œë“œ ë‚ ì§œ
category_ids = [] #ë™ì˜ìƒ ì¹´í…Œê³ ë¦¬ id
views = [] #ì¡°íšŒìˆ˜
likes = [] #ì¢‹ì•„ìš” ìˆ˜
dislikes = [] #ì‹«ì–´ìš” ìˆ˜
comments = [] #ëŒ“ê¸€ ìˆ˜
tags = [] #íƒœê·¸ ì •ë³´

hours = [] #ë™ì˜ìƒ ì¬ìƒê¸¸ì´(ì‹œ)
mins = [] #ë™ì˜ìƒ ì¬ìƒê¸¸ì´(ë¶„)
secs = [] #ë™ì˜ìƒ ì¬ìƒê¸¸ì´(ì´ˆ)

#ë¦¬ìŠ¤íŠ¸ í˜•íƒœì˜ tags ì •ë³´ë¥¼ ì¼ë ¨ì˜ ë¬¸ìì—´ë¡œ ë³€í™˜í•´ì£¼ëŠ” í•¨ìˆ˜
def Tags2Strings(tags_list):
    
    string = ""
    for i in range(len(tags_list)):
        string = string + tags_list[i] + "|"
    string = string[:-1] #ë¬¸ìì—´ ë§¨ ëì— ë¶™ì€ "|"ì€ ì—†ì• ë²„ë¦¼
    
    return string

#ë°ì´í„° ì¶”ì¶œ
for i in range(len(item_itzy)): #ITZY
    request = youtube.videos().list(
        id = item_itzy["video_id"][i],
        part = "snippet,contentDetails,statistics"
    )
    video_response = request.execute()
    
    if video_response["items"] == []: #ë™ì˜ìƒ ì •ë³´ê°€ ì—†ì„ ê²½ìš° '-'ê°’ ì…ë ¥
        titles.append("-")
        ids.append("-")
        dates.append("-")
        category_ids.append("-")
        views.append("-")
        likes.append("-")
        dislikes.append("-")
        comments.append("-")
        tags.append("-")
        hours.append("-")
        mins.append("-")
        secs.append("-")
    
    else:
        titles.append(video_response["items"][0]["snippet"]["title"])
        ids.append(video_response["items"][0]["id"])
        dates.append(video_response["items"][0]["snippet"]["publishedAt"].split("T")[0])
        category_ids.append(video_response["items"][0]["snippet"]["categoryId"])
        
        #views ë°ì´í„°
        if "viewCount" in video_response["items"][0]["statistics"].keys():
            views.append(video_response["items"][0]["statistics"]["viewCount"])
            
        else:
            views.append("-")
        
        #comments ë°ì´í„°
        if "commentCount" in video_response["items"][0]["statistics"].keys():
            comments.append(video_response["items"][0]["statistics"]["commentCount"])
            
        else:
            comments.append("-")
        
        #likes ë°ì´í„°
        if "likeCount" in video_response["items"][0]["statistics"].keys():
            likes.append(video_response["items"][0]["statistics"]["likeCount"])
            
        else: #likeCountê°€ ì—†ëŠ” ê²½ìš°ë„ ìˆìŒ
            likes.append("-")
        
        #dislikes ë°ì´í„°
        if "dislikeCount" in video_response["items"][0]["statistics"].keys():
            dislikes.append(video_response["items"][0]["statistics"]["dislikeCount"])
            
        else:
            dislikes.append("-")
        
        #tags ë°ì´í„°
        if "tags" in video_response["items"][0]["snippet"].keys():
            tags_string = video_response["items"][0]["snippet"]["tags"]
            tags.append(Tags2Strings(tags_string))
            
        else: #tagsê°€ ì—†ëŠ” ê²½ìš°ë„ ìˆìŒ(shorts í˜•ì‹ì˜ ì˜ìƒë“¤)
            tags.append("-")
        
        #hour, min, sec ë°ì´í„°
        duration = re.findall(r'\d+', video_response["items"][0]["contentDetails"]["duration"])
        if len(duration) == 3:
            hours.append(duration[0])
            mins.append(duration[1])
            secs.append(duration[2])
        elif len(duration) == 2:
            hours.append("-")
            mins.append(duration[0])
            secs.append(duration[1])
        else:
            hours.append("-")
            mins.append("-")
            secs.append(duration[0])
        
        
video_df = pd.DataFrame({"title": titles, "videoId": ids, "date": dates, "categoryId": category_ids,
                        "view": views, "like": likes, "dislike": dislikes, "comment": comments, "tags": tags,
                        "hour": hours, "min": mins, "sec": secs})
video_df
```
