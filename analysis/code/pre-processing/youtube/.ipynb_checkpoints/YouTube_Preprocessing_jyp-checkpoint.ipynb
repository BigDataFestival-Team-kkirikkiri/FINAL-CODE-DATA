# (2) 소속사 채널 동영상 데이터
video_jyp 데이터프레임에서 뮤직비디오 영상 데이터만 추출

```python
#결측치 확인
video_jyp.info()
```

## (2-1) 분석 기간 내의 데이터 추출 및 날짜 타입 변경
2023년 1월 ~ 2023년 6월의 동영상 데이터만 추출한 후, 'date'를 문자형에서 날짜 타입인 datetime으로 변경함.

```python
#날짜 변수인 'date'는 문자형임
video_jyp["date"]
```

```python
index_list = [] #조건에 해당하는 행의 index list

for i in range(len(video_jyp)):
    if int(video_jyp["date"][i].split("-")[0]) == 2023: #연도가 2023년인 경우
        if 1 <= int(video_jyp["date"][i].split("-")[1]) <= 6: #1월 ~ 6월인 경우
            index_list.append(i)

data_jyp = video_jyp.iloc[index_list].reset_index(drop=True)
data_jyp
```

```python
#date변수를 날짜 타입인 datetime으로 변환
for i in range(len(data_jyp)):
    data_jyp["date"][i] = datetime.datetime.strptime(data_jyp["date"][i], "%Y-%m-%d")
    
data_jyp
```

## (2-2) 뮤직비디오 영상 데이터 추출
영상 제목에 'M/V'라는 단어가 있는 영상 데이터만 추출

```python
index_list = []

for i in range(len(data_jyp)):
    if ("M/V" in data_jyp["title"][i].split()): #제목에 "M/V"가 포함된 경우
        index_list.append(i)

data_jyp = data_jyp.iloc[index_list].reset_index(drop=True)
data_jyp
```

```python
#data_jyp를 csv파일로 저장
data_jyp.to_csv("../../../data/YouTube/data_jyp.csv", index=False)
```
