## (2-1) Î∂ÑÏÑù Í∏∞Í∞Ñ ÎÇ¥Ïùò Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
2023ÎÖÑ 1Ïõî ~ 2023ÎÖÑ 6ÏõîÏùò ÎèôÏòÅÏÉÅ Îç∞Ïù¥ÌÑ∞Îßå Ï∂îÏ∂ú

```python
#ÎÇ†Ïßú Î≥ÄÏàòÏù∏ 'date'Îäî Î¨∏ÏûêÌòïÏûÑ
video_yg["date"]
```

```python
index_list = [] #Ï°∞Í±¥Ïóê Ìï¥ÎãπÌïòÎäî ÌñâÏùò index list

for i in range(len(video_yg)):
    if int(video_yg["date"][i].split("-")[0]) == 2023: #Ïó∞ÎèÑÍ∞Ä 2023ÎÖÑÏù∏ Í≤ΩÏö∞
        if 1 <= int(video_yg["date"][i].split("-")[1]) <= 6: #1Ïõî ~ 6ÏõîÏù∏ Í≤ΩÏö∞
            index_list.append(i)

data_yg = video_yg.iloc[index_list].reset_index(drop=True)
data_yg
```

## (2-2) ÎÆ§ÏßÅÎπÑÎîîÏò§ ÏòÅÏÉÅ Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
ÏòÅÏÉÅ Ï†úÎ™©Ïóê 'M/V'ÎùºÎäî Îã®Ïñ¥Í∞Ä ÏûàÎäî ÏòÅÏÉÅ Îç∞Ïù¥ÌÑ∞Îßå Ï∂îÏ∂ú

```python
index_list = []

for i in range(len(data_yg)):
    if ("M/V" in data_yg["title"][i].split()): #Ï†úÎ™©Ïóê "M/V"Í∞Ä Ìè¨Ìï®Îêú Í≤ΩÏö∞
        index_list.append(i)

data_yg = data_yg.iloc[index_list].reset_index(drop=True)
data_yg
```

### (1Î≤à) video_blackpink
Î∏îÎûôÌïëÌÅ¨

```python
index_list = [] #Ï°∞Í±¥Ïóê Ìï¥ÎãπÌïòÎäî ÌñâÏùò index list

for i in range(len(video_blackpink)):
    if int(video_blackpink["date"][i].split("-")[0]) == 2023: #Ïó∞ÎèÑÍ∞Ä 2023ÎÖÑÏù∏ Í≤ΩÏö∞
        if 1 <= int(video_blackpink["date"][i].split("-")[1]) <= 6: #1Ïõî ~ 6ÏõîÏù∏ Í≤ΩÏö∞
            index_list.append(i)

data_blackpink = video_blackpink.iloc[index_list].reset_index(drop=True)
data_blackpink
```

```python
#dateÎ≥ÄÏàòÎ•º ÎÇ†Ïßú ÌÉÄÏûÖÏù∏ datetimeÏúºÎ°ú Î≥ÄÌôò
for i in range(len(data_blackpink)):
    data_blackpink["date"][i] = datetime.datetime.strptime(data_blackpink["date"][i], "%Y-%m-%d")
    
data_blackpink
```
