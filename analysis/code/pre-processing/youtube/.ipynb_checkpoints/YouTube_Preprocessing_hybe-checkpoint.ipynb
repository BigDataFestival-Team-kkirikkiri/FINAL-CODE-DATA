## (2-1) ë¶„ì„ ê¸°ê°„ ë‚´ì˜ ë°ì´í„° ì¶”ì¶œ ë° ë‚ ì§œ íƒ€ì… ë³€ê²½
2023ë…„ 1ì›” ~ 2023ë…„ 6ì›”ì˜ ë™ì˜ìƒ ë°ì´í„°ë§Œ ì¶”ì¶œí•œ í›„, 'date'ë¥¼ ë¬¸ìí˜•ì—ì„œ ë‚ ì§œ íƒ€ì…ì¸ datetimeìœ¼ë¡œ ë³€ê²½í•¨.

```python
#ë‚ ì§œ ë³€ìˆ˜ì¸ 'date'ëŠ” ë¬¸ìí˜•ì„
video_hybe["date"]
```

```python
index_list = [] #ì¡°ê±´ì— í•´ë‹¹í•˜ëŠ” í–‰ì˜ index list

for i in range(len(video_hybe)):
    if int(video_hybe["date"][i].split("-")[0]) == 2023: #ì—°ë„ê°€ 2023ë…„ì¸ ê²½ìš°
        if 1 <= int(video_hybe["date"][i].split("-")[1]) <= 6: #1ì›” ~ 6ì›”ì¸ ê²½ìš°
            index_list.append(i)

data_hybe = video_hybe.iloc[index_list].reset_index(drop=True)
data_hybe
```

## (2-2) ë®¤ì§ë¹„ë””ì˜¤ ì˜ìƒ ë°ì´í„° ì¶”ì¶œ
ì˜ìƒ ì œëª©ì— 'Official'ê³¼ 'MV'ë¼ëŠ” ë‹¨ì–´ê°€ ìˆëŠ” ì˜ìƒ ë°ì´í„°ë§Œ ì¶”ì¶œ

```python
index_list = []

for i in range(len(data_hybe)):
    if ("Official" in data_hybe["title"][i].split()) and ("MV" in data_hybe["title"][i].split()): #ì œëª©ì— "Official"ê³¼ "MV"ê°€ í¬í•¨ëœ ê²½ìš°
        index_list.append(i)

data_hybe = data_hybe.iloc[index_list].reset_index(drop=True)
data_hybe
```

```python
#dateë³€ìˆ˜ë¥¼ ë‚ ì§œ íƒ€ì…ì¸ datetimeìœ¼ë¡œ ë³€í™˜
for i in range(len(data_hybe)):
    data_hybe["date"][i] = datetime.datetime.strptime(data_hybe["date"][i], "%Y-%m-%d")
    
data_hybe
```

```python
#data_hybeë¥¼ csvíŒŒì¼ë¡œ ì €ì¥
data_hybe.to_csv("../../../data/YouTube/data_hybe.csv", index=False)
```
