video_sm 데이터프레임에서 뮤직비디오 영상 데이터, 티저 데이터만 추출

## (2-1) 분석 기간 내의 데이터 추출
2023년 1월 ~ 2023년 6월의 동영상 데이터만 추출

```python
#날짜 변수인 'date'는 문자형임
video_sm["date"]
```

```python
index_list = [] #조건에 해당하는 행의 index list

for i in range(len(video_sm)):
    if int(video_sm["date"][i].split("-")[0]) == 2023: #연도가 2023년인 경우
        if 1 <= int(video_sm["date"][i].split("-")[1]) <= 6: #1월 ~ 6월인 경우
            index_list.append(i)

data_sm = video_sm.iloc[index_list].reset_index(drop=True)
data_sm
```

```python
data_sm.head()
```

- SM은 뮤직비디오는 끝에 MV 가 붙고, 티저는 MV Teaser 가 붙는다.

## (2-2) 뮤직비디오 영상 데이터 추출
- 영상 제목에'MV'라는 단어가 있는 영상 데이터만 추출
- MV 라는 단어가 있는 영상 데이터를 뽑으면 MV Teaser 도 자연스럽게 추출되기 떄문임

```python
index_list = []

for i in range(len(data_sm)):
    if ("MV" in data_sm["title"][i].split()):
        index_list.append(i)

data_sm = data_sm.iloc[index_list].reset_index(drop=True)
data_sm
```
