## (2-1) 분석 기간 내의 데이터 추출
2023년 1월 ~ 2023년 6월의 동영상 데이터만 추출

```python
#날짜 변수인 'date'는 문자형임
video_yg["date"]
```

```python
index_list = [] #조건에 해당하는 행의 index list

for i in range(len(video_yg)):
    if int(video_yg["date"][i].split("-")[0]) == 2023: #연도가 2023년인 경우
        if 1 <= int(video_yg["date"][i].split("-")[1]) <= 6: #1월 ~ 6월인 경우
            index_list.append(i)

data_yg = video_yg.iloc[index_list].reset_index(drop=True)
data_yg
```

## (2-2) 뮤직비디오 영상 데이터 추출
영상 제목에 'M/V'라는 단어가 있는 영상 데이터만 추출

```python
index_list = []

for i in range(len(data_yg)):
    if ("M/V" in data_yg["title"][i].split()): #제목에 "M/V"가 포함된 경우
        index_list.append(i)

data_yg = data_yg.iloc[index_list].reset_index(drop=True)
data_yg
```

## (3-1) 분석 기간 내의 데이터 추출 및 날짜 타입 변경
2023년 1월 ~ 2023년 6월의 동영상 데이터만 추출한 후, 'date'를 문자형에서 날짜 타입인 datetime으로 변경함.

### (1번) video_blackpink
블랙핑크

```python
index_list = [] #조건에 해당하는 행의 index list

for i in range(len(video_blackpink)):
    if int(video_blackpink["date"][i].split("-")[0]) == 2023: #연도가 2023년인 경우
        if 1 <= int(video_blackpink["date"][i].split("-")[1]) <= 6: #1월 ~ 6월인 경우
            index_list.append(i)

data_blackpink = video_blackpink.iloc[index_list].reset_index(drop=True)
data_blackpink
```

```python
#date변수를 날짜 타입인 datetime으로 변환
for i in range(len(data_blackpink)):
    data_blackpink["date"][i] = datetime.datetime.strptime(data_blackpink["date"][i], "%Y-%m-%d")
    
data_blackpink
```

```python
#data_blackpink를 csv파일로 저장
data_blackpink.to_csv("../../../data/YouTube/data_blackpink.csv", index=False)
```
