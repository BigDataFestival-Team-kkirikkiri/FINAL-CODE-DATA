```python
index_list = [] #ì¡°ê±´ì— í•´ë‹¹í•˜ëŠ” í–‰ì˜ index list

for i in range(len(video_jyp)):
    if int(video_jyp["date"][i].split("-")[0]) == 2023: #ì—°ë„ê°€ 2023ë…„ì¸ ê²½ìš°
        if 1 <= int(video_jyp["date"][i].split("-")[1]) <= 6: #1ì›” ~ 6ì›”ì¸ ê²½ìš°
            index_list.append(i)

data_jyp = video_jyp.iloc[index_list].reset_index(drop=True)
data_jyp
```
