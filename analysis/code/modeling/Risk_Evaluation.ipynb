# (2) 등락률 계산
- 날짜별 등락률을 계산함.
- 첫 번째 행(2023-01-02)은 전일종가 정보가 없으므로 등락률 계산이 불가하므로 등락률을 '0'으로 처리함.

$$등락률 = \frac{현재가 - 전일종가}{전일종가}×100 $$
$$≒ \frac{현재 predicted - 전일 actual}{전일 actual}×100$$

```python
#등락률을 계산해주는 함수 생성
def fluctuation_rate(data):
    
    ratio_list = []
    for i in range(len(data)):
        if i == 0: #첫 번째 행의 경우
            ratio = 0
        else:
            ratio = (data["predicted"][i] - data["actual"][i-1]) / data["actual"][i-1] * 100
    
        ratio_list.append(ratio)
    
    return ratio_list
```

```python
hybe["ratio"] = fluctuation_rate(hybe)
jyp["ratio"] = fluctuation_rate(jyp)
sm["ratio"] = fluctuation_rate(sm)
yg["ratio"] = fluctuation_rate(yg)
```

# (3) 위험도 분류
- 3%, 5%, 10% 기준으로 위험도 분류함.
- 등락률을 계산이 불가하여 '0'으로 처리한 첫 번째 행은 위험도도 '0'으로 처리함.
- 안전: (당일 등락률) >= (기준 퍼센트)
- 보통: -(기준 퍼센트) < (당일 등락률) < (기준 퍼센트)
- 위험: (당일 등락률) <= -(기준 퍼센트)

```python
#3% 기준으로 위험도 분류하는 함수 생성
def risk3(data):
    
    risk_list = []
    for i in range(len(data)):
        if i == 0: #첫 번째 행의 경우
            risk = 0
        else:
            ratio = data["ratio"][i]
            if ratio >= 3:
                risk = "안전"
            elif -3 < ratio < 3:
                risk = "보통"
            elif ratio <= -3:
                risk = "위험"
                
        risk_list.append(risk)
    return risk_list

####################################################

#5% 기준으로 위험도 분류하는 함수 생성
def risk5(data):
        
    risk_list = []
    for i in range(len(data)):
        if i == 0: #첫 번째 행의 경우
            risk = 0
        else:
            ratio = data["ratio"][i]
            if ratio >= 5:
                risk = "안전"
            elif -5 < ratio < 5:
                risk = "보통"
            elif ratio <= -5:
                risk = "위험"
                
        risk_list.append(risk)
    return risk_list

#####################################################

#10% 기준으로 위험도 분류하는 함수 생성
def risk10(data):
        
    risk_list = []
    for i in range(len(data)):
        if i == 0: #첫 번째 행의 경우
            risk = 0
        else:
            ratio = data["ratio"][i]
            if ratio >= 10:
                risk = "안전"
            elif -10 < ratio < 10:
                risk = "보통"
            elif ratio <= -10:
                risk = "위험"
                
        risk_list.append(risk)
    return risk_list
```

```python
hybe["risk3"] = risk3(hybe)
hybe["risk5"] = risk5(hybe)
hybe["risk10"] = risk10(hybe)

jyp["risk3"] = risk3(jyp)
jyp["risk5"] = risk5(jyp)
jyp["risk10"] = risk10(jyp)

sm["risk3"] = risk3(sm)
sm["risk5"] = risk5(sm)
sm["risk10"] = risk10(sm)

yg["risk3"] = risk3(yg)
yg["risk5"] = risk5(yg)
yg["risk10"] = risk10(yg)
```
